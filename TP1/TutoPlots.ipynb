```python
#This is a notebook to illustrate some basic functionalities of the matplotlib package to generate figures.
#For a complete documentation, you can refer to the official website https://matplotlib.org

import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
from matplotlib import collections  as mc


# Fixing random state for reproducibility
np.random.seed(1234)

ub= 20
lb=1

x = np.arange(lb, ub, 0.2)
#Number of steps
#print(*x.shape) 

data =  - x**3 + 2*0 + 10000 + np.random.rand(*x.shape) * 3000  

over = data 
good = - x**3 + 11500
under = - x**2 + 10500


#Only original
plt.figure(0)
#scatter makes dots 
plt.scatter(x, data,  c="g", alpha=1, marker="*",
            label="Original Data")

plt.xlabel("Distance To city")
plt.ylabel("Price")
plt.legend(loc='upper right')
plt.title ("House prices per Meter Square")

#Size of the window
plt.rcParams["figure.figsize"] = (8,4)
#Save Plot
plt.savefig('original.png')
plt.show()


```

```python
plt.figure(1)
#scatter makes dots 
plt.scatter(x, data,  c="g", alpha=1, marker="*",
            label="Original Data")

#plot make things connected
plt.plot(x, over,  c="r", alpha=0.6,
           label="Overfitting Model")

plt.xlabel("Distance To city")
plt.ylabel("Price")
plt.legend(loc='upper right')
plt.title ("House prices per Meter Square: Overfitting")

#Size of the window
plt.rcParams["figure.figsize"] = (8,4)
#Save Plot
plt.savefig('overfitting.png')
plt.show()

```

```python
plt.figure(2)
#scatter makes dots 
plt.scatter(x, data,  c="g", alpha=1, marker="*",
            label="Original Data")

#Continious line
plt.plot(x, under,  c="y", alpha=0.6, 
            label="Underfitting function")

plt.xlabel("Distance To city")
plt.ylabel("Price")
plt.legend(loc='upper right')
plt.title ("House prices per Meter Square: Underfitting")

#Size of the window
plt.rcParams["figure.figsize"] = (8,4)
#plt.savefig('books_read.png')
plt.savefig('underfitting.png')
plt.show()
```

```python
plt.figure(4)
#scatter makes dots 
plt.scatter(x, data,  c="g", alpha=1, marker="*",
            label="Original Data")


plt.plot(x, good , c="b", alpha=0.6, 
           label="Good fit")

plt.xlabel("Distance To city")
plt.ylabel("Price")
plt.legend(loc='upper right')
plt.title ("House prices per Meter Square")

#Size of the window
plt.rcParams["figure.figsize"] = (8,4)
#plt.savefig('books_read.png')
plt.savefig('goodfit.png')
plt.show()
```

```python
##Overfitting with Decision Trees
plt.figure(5)
x = [3,5,7,9,11,13,15]
ytest = [0.68, 0.8, 0.84, 0.82, 0.77, 0.69 ,0.58]
ytrain = [0.7, 0.77, 0.82, 0.86, 0.90, 0.92 ,0.95]

plt.plot(x, ytest,  c="r", 
         label="Testing Accuracy")

plt.plot(x, ytrain,  c="b", 
         label="Training Accuracy")

plt.xlabel("Tree Depth")
plt.ylabel("Accuracy")
plt.legend(loc='upper left')
plt.title ("Overfitting with Decision Trees")

plt.rcParams["figure.figsize"] = (8,4)
plt.savefig('treeoverfitting.png')

plt.show()
```

```python
##Overfitting: simple to complex models

plt.figure(6)
x = [3,5,7,9,11,13,15]
ytest = [0.68, 0.8, 0.84, 0.82, 0.77, 0.69 ,0.58]
ytrain = [0.7, 0.77, 0.82, 0.86, 0.90, 0.92 ,0.95]

plt.plot(x, ytest,  c="r", 
         label="Complex model: Testing Accuracy")

plt.plot(x, ytrain,  c="b", 
         label="Complex model: Training Accuracy")

plt.xlabel("Complexity of the Model (left = simple, right = too complex)")
plt.ylabel("Accuracy")
plt.legend(loc='upper left')
plt.title ("Overfitting illustration with Complex Models")

plt.rcParams["figure.figsize"] = (8,4)

#get current axes
ax = plt.gca()
ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])

#plt.grid(True)


plt.savefig('exampleoverfitting.png')

plt.show()
```
